#### Лемматизация

```python
# Переведём тексты в стандартный для Python формат: сделаем кодировку Unicode 
corpus = data['text'].values

# Проверка
corpus
```

```python
# Загрузка необходимых данных
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
# Создаем объект lemmatizer, который будет использоваться для лемматизации каждого слова
lemmatizer = WordNetLemmatizer()
```

```python
stop_words = set(stopwords.words('english'))  # Получаем набор стоп-слов на английском
```

```python
# Функция для преобразования POS-тегов в формат WordNet
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
# Функция для очистки токенов от знаков препинания и лишних символов
def clean_token(token):
    return re.sub(r'[^a-zA-Z0-9]', '', token)
```

```python
# Лемматизация текста с устранением стоп-слов и знаков препинания
def lemmatize_sentence(sentence):
    # Токенизация текста
    tokens = nltk.word_tokenize(sentence)
    # Убираем стоп-слова и знаки препинания, затем лемматизируем оставшиеся токены
    lemmas = [lemmatizer.lemmatize(clean_token(token), get_wordnet_pos(token)) 
              for token in tokens if clean_token(token).lower() not in stop_words and clean_token(token)]
    return lemmas
```

```python
# Применение лемматизации к каждому элементу массива
lemmatized_data = np.array([lemmatize_sentence(sentence) for sentence in corpus])

# Вывод результата (первые 5 примеров)
for i, (original, lemmatized) in enumerate(zip(corpus, lemmatized_data)):
    if i < 5: 
        print(f"Исходный текст: {original}")
        print(f"Преобразованный текст: {' '.join(lemmatized)}\n\n")
```

Инициализируем токенизатор как объект класса BertTokenizer(). Передадим ему аргумент vocab_file — это файл со словарём, на котором обучалась модель.

Преобразуем текст в номера токенов из словаря методом encode()

```python
# Токенизация с паддингом и усечением
tokenized = data['text'].apply(
    lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True, padding='max_length', max_length=512)
)

tokenized_array = np.array(tokenized.tolist())

# Проверка
print(f'Максимальная длина токенизированных текстов: {tokenized_array.shape[1]}')
print(f'Токенизированные тексты:\n {tokenized_array}')
```

Создадим маску для действительно важных токенов.

```python
attention_mask = np.where(tokenized_array != 0, 1, 0)

# Вывод маски
attention_mask
```

### Эмбеддинги BERT

Начнём преобразование текстов в эмбеддинги. 

```python
config = transformers.AutoModel.from_pretrained('unitary/toxic-bert')
model = transformers.AutoModel.from_pretrained('unitary/toxic-bert', config=config)
```

```python
batch_size = 200
embeddings = []
for i in notebook.tqdm(range(tokenized_array.shape[0] // batch_size)):
        batch = torch.LongTensor(tokenized_array[batch_size*i:batch_size*(i+1)]) 
        attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)])
        
        with torch.no_grad():
            batch_embeddings = model(batch, attention_mask=attention_mask_batch)
        
        embeddings.append(batch_embeddings[0][:,0,:].numpy())
```

Соберём все эмбеддинги в матрицу признаков.

```python
features = np.concatenate(embeddings)

# Проверка
print(f'Размер матрицы признаков: {features.shape}')
print(f'Матрица признаков:\n {features}')
```

## Обучение

### Разделение на тренировочную и тестовую выборки

```python
X_train, X_test, y_train, y_test = train_test_split(features, data['toxic'], random_state=RANDOM_STATE, stratify=data['toxic'])
```

### Использование пайплайна

```python
# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5)
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5)
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 5) 
    },
    
    # словарь для модели SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE)],
        'models__C': range(1, 2)
    },
    
    # словарь для модели LGBMClassifier()
    {
        'models': [LGBMClassifier(random_state=RANDOM_STATE, objective='regression', metric='rmse')],
        'models__max_depth': [1, 5],  
        'models__learning_rate': [0.01, 0.1, 1.0],
    }
]
```

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='f1',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print('Метрика лучшей модели на кросс-валидацинной выборке:', randomized_search.best_score_)
```

### Тестирование

```python
best_model = randomized_search.best_estimator_
pred = best_model.predict(X_test)
print(f'f1-score на тестовой выборке: {f1_score(y_test, pred)}')
```
